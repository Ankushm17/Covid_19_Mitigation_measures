Importing the data of confirmed cases, deaths, recovered (Jonh hopkins CSSEGIS) data uptil 20th Oct 2020
Changing the columns to rows and sorting the values so as to rearrange the data according to our favour

```
confirmed_data_path="/content/drive/My Drive/Data/time_series_covid19_confirmed_global.csv"
confirmed_data = pd.read_csv(confirmed_data_path, index_col="Country/Region")
death_data_path = "/content/drive/My Drive/Data/time_series_covid19_deaths_global.csv"
death_data = pd.read_csv(death_data_path, index_col="Country/Region")
recovered_data_path = "/content/drive/My Drive/Data/time_series_covid19_recovered_global.csv"
recovered_data = pd.read_csv(recovered_data_path, index_col="Country/Region")
```

```
confirmed_data=confirmed_data.drop(['Province/State','Lat','Long'], axis = 1) 
death_data=death_data.drop(['Province/State','Lat','Long'], axis = 1) 
recovered_data=recovered_data.drop(['Province/State','Lat','Long'], axis = 1) 
```

```
confirmed_data=confirmed_data.reset_index()
```

```
confirmed_data=confirmed_data.melt(id_vars=["Country/Region"], var_name="Date", value_name="Confirmed_cases")
```

```
death_data=death_data.melt(id_vars=["Country/Region"], var_name="Date", value_name="Confirmed_deaths")
```

```
recovered_data=recovered_data.melt(id_vars=["Country/Region"], var_name="Date", value_name="Recovered_cases")
```

```
recovered_data=recovered_data.sort_values(["Country/Region","Date"])
```

We need to concatenate the data of deaths, confirmed, recovered.

```
concat_data=pd.concat([confirmed_data, death_data, recovered_data])
```
